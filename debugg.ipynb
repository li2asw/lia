```python
from unicodedata import normalize
import pandas as pd
```

```python
def pt_normalize(word):
        return normalize("NFKD", word).encode("ASCII", "ignore").decode("ASCII")
```

```python
df = pd.read_csv("all_nouns.csv")
```

```python
df.head(10)
```

```python
df['noun'] = df['noun'].apply(pt_normalize)
df['support'] = df['support'].apply(pt_normalize)
```

```python
dict = {'noun': ['desenvolvedor', 'desenvolvedora', 'desenvolvedores', 'desenvolvedoras'], 
        'support': ['desenvolvedor', 'desenvolvedor', 'desenvolvedor', 'desenvolvedor'], 
        'len': [13, 14, 15, 15],
        'variation': [4, 4, 4, 4],
       }

df = df.append(dict, ignore_index = True)
```
